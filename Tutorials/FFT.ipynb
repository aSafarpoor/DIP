### imports

```python
import matplotlib.pyplot as plt
import numpy as np
from time import sleep

from scipy.fft import fft, fftfreq

import cv2
```

```python
from google.colab.patches import cv2_imshow
```

### part one: simple fft on sin()

simple sine:

```python
x = np.arange(0,4*np.pi,0.1)
y = np.sin(x)

plt.plot(x,y)
plt.show()
```

```python
F = fft(y)
plt.plot(x,F)
plt.show()
```

what it means?
It means we had sine with size 1 and in image coefficients are 1 too.

```python
F = fft(y)
mag = np.abs(F)
fig, axs = plt.subplots(2)

axs[0].plot(x,y)
axs[1].plot(x,mag)

plt.show()
```

```python
x = np.arange(0,4*np.pi,0.1)
y = np.sin(x) + 2*np.sin(2*x) + 0.5*np.sin(4*x) + 3*np.sin(0.2+x)

F = fft(y)

fig, axs = plt.subplots(5)
axs[0].plot(x,y)
axs[1].plot(x,np.abs(F))
axs[2].plot(x,np.angle(F))
axs[3].plot(x,np.real(F))
axs[4].plot(x,np.imag(F))

plt.show()
```

### part two: fft on image

```python
path = 'download.jfif'
img_color = cv2.imread(path)
img = cv2.imread(path,cv2.IMREAD_GRAYSCALE)

try:
    cv2.imshow('img',img)
except:
    cv2_imshow(img)
cv2_imshow(img_color)
cv2.waitKey(10)
```

```python
print(img.shape)
```

```python
F = np.fft.fft2(img)
print(F.shape)
print(F.dtype)
```

```python
F = np.fft.fft2(img)
# F = np.fft.fftshift(F)

mag = np.abs(F)
magd = 255 * mag / max(map(max,mag))
logmag = np.log(magd)
cv2_imshow(img)
cv2_imshow(magd)
cv2_imshow(logmag)
```
